# Exploratory Data Analysis Notebook

This notebook explores government spending datasets to uncover patterns, trends, and insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load datasets
transactions = pd.read_csv('../datasets/raw/transactions.csv')
schemes = pd.read_csv('../datasets/raw/schemes.csv')

print(f"Transactions shape: {transactions.shape}")
print(f"Schemes shape: {schemes.shape}")
transactions.head()
```

```python
# Basic statistics
print('Transactions Info:')
print(transactions.info())
print('\nSchemes Info:')
print(schemes.info())
print('\nTransactions describe:')
print(transactions.describe())
print('\nSchemes describe:')
print(schemes.describe())
```

```python
# Spending by department
dept_spending = transactions.groupby('department')['amount'].sum().sort_values(ascending=False)
print('Spending by department:')
print(dept_spending)

# Plotting
plt.figure(figsize=(10, 6))
dept_spending.plot(kind='barh')
plt.title('Total Spending by Department')
plt.xlabel('Amount (INR)')
plt.tight_layout()
plt.show()
```

```python
# Monthly spending trends
transactions['date'] = pd.to_datetime(transactions['date'])
transactions['month'] = transactions['date'].dt.to_period('M')
monthly_spending = transactions.groupby('month')['amount'].sum()
print('Monthly spending:')
print(monthly_spending)

plt.figure(figsize=(10, 4))
monthly_spending.plot(marker='o')
plt.title('Monthly Spending Trend')
plt.xlabel('Month')
plt.ylabel('Amount (INR)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
